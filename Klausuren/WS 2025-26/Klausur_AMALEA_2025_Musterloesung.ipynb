Hier finden Sie die Musterlösungen zu allen Multiple-Choice- und offenen Fragen der Klausur.

## Musterlösung: Multiple-Choice-Fragen

**Frage 1:** In welcher Phase des QUA³CK-Prozessmodells wird Data Leakage am effektivsten verhindert?

**Richtige Antwort:** Acquire & Clean

Begründung: In dieser Phase werden Datenquellen geprüft, bereinigt und korrekt getrennt, um Data Leakage zu vermeiden.

**Frage 2:** Was ist der Unterschied zwischen Overfitting und Underfitting bei Machine Learning Modellen?

**Richtige Antwort:** Overfitting: Modell passt sich zu stark an Trainingsdaten an; Underfitting: Modell kann Muster nicht erfassen

Begründung: Overfitting = Überanpassung, Underfitting = zu geringe Modellkomplexität.

**Frage 3:** Welche der folgenden Aufgaben gehört NICHT zur Explorativen Datenanalyse (EDA)?

**Richtige Antwort:** Hyperparameter-Tuning

Begründung: Hyperparameter-Tuning ist Teil der Modelloptimierung, nicht der EDA.

**Frage 4:** Was ist ein typisches Problem einzelner Entscheidungsbäume im Machine Learning?

**Richtige Antwort:** Sie neigen zu Overfitting

Begründung: Einzelne Entscheidungsbäume sind sehr flexibel und können sich zu stark an Trainingsdaten anpassen.

**Frage 5:** Welche Maßnahme trägt am meisten zur Reproduzierbarkeit von Data-Science-Projekten bei?

**Richtige Antwort:** Verwendung von Docker-Containern

Begründung: Docker garantiert reproduzierbare Umgebungen.

**Frage 6:** Warum ist Feature Engineering im Machine Learning wichtig?

**Richtige Antwort:** Feature Engineering kann die Modellgüte signifikant verbessern

Begründung: Neue oder optimierte Features können die Leistung deutlich steigern.

**Frage 7:** Welches Verfahren wird typischerweise zur Überprüfung der Robustheit eines Modells gegenüber Ausreißern eingesetzt?

**Richtige Antwort:** Cross-Validation

Begründung: Cross-Validation prüft die Robustheit gegenüber Ausreißern und Varianz.

**Frage 8:** Was ist ein typisches Symptom für Data Leakage in einem Machine Learning Projekt?

**Richtige Antwort:** Sehr hohe Trainings- und Testgenauigkeit, aber schlechte Generalisierung auf neue Daten

Begründung: Data Leakage führt zu zu guten Ergebnissen auf bekannten Daten, aber schlechter Generalisierung.

**Frage 9:** Welches Tool wird im Kurs als zentrales Werkzeug für das Tracking von ML-Experimenten verwendet?

**Richtige Antwort:** MLflow

Begründung: MLflow ist das zentrale Tool für Experiment- und Modelltracking im Kurs.

**Frage 10:** Warum sind Random Forests robuster als einzelne Entscheidungsbäume?

**Richtige Antwort:** Sie reduzieren Overfitting durch Aggregation vieler Bäume

Begründung: Random Forests mitteln viele Bäume und sind dadurch robuster.

**Frage 11:** Was unterscheidet das KNN-Verfahren grundlegend von modellbasierten Algorithmen?

**Richtige Antwort:** KNN ist ein instanzbasiertes Verfahren und speichert alle Trainingsdaten

Begründung: KNN benötigt keine Modellparameter, sondern nutzt die Trainingsdaten direkt.

**Frage 12:** Wofür wird die Hauptkomponentenanalyse (PCA) im Machine Learning eingesetzt?

**Richtige Antwort:** Dimensionsreduktion und Visualisierung

Begründung: PCA reduziert die Anzahl der Merkmale und hilft bei der Visualisierung.

**Frage 13:** Warum ist die Standardisierung der Features beim KNN-Algorithmus besonders wichtig?

**Richtige Antwort:** Sie ist wichtig, da KNN auf Distanzmaßen basiert

Begründung: Ohne Standardisierung dominieren Features mit großem Wertebereich.

**Frage 14:** Was ist der Hauptvorteil von Cross-Validation bei der Modellbewertung?

**Richtige Antwort:** Cross-Validation reduziert die Varianz der Modellbewertung

Begründung: Durch Mittelung über mehrere Splits wird die Bewertung stabiler.

**Frage 15:** Was ist eine Einschränkung linearer Modelle im Machine Learning?

**Richtige Antwort:** Sie können nicht-lineare Zusammenhänge nicht abbilden

Begründung: Lineare Modelle sind auf lineare Beziehungen beschränkt.

**Frage 16:** Wie kann Overfitting bei Deep-Learning-Modellen im Bereich Computer Vision reduziert werden?

**Richtige Antwort:** Data Augmentation kann helfen, Overfitting zu reduzieren

Begründung: Durch künstliche Erweiterung des Datensatzes wird das Modell robuster.

**Frage 17:** Was ist ein Vorteil von Transfer Learning gegenüber dem Training eines Modells von Grund auf?

**Richtige Antwort:** Geringerer Datenbedarf und schnellere Konvergenz

Begründung: Vortrainierte Modelle benötigen weniger Daten und lernen schneller.

**Frage 18:** Was ist bei der Arbeit mit Hyperparametern in Machine Learning Modellen zu beachten?

**Richtige Antwort:** Hyperparameter müssen vor dem Training festgelegt und ggf. optimiert werden

Begründung: Sie werden nicht automatisch gelernt, sondern müssen gewählt werden.

**Frage 19:** Was ist ein häufiger Fehler beim Hyperparameter-Tuning, der zu einer Überschätzung der Modellgüte führen kann?

**Richtige Antwort:** Nutzung der Testdaten für Hyperparameter-Tuning

Begründung: Die Testdaten dürfen erst zur finalen Bewertung genutzt werden.

**Frage 20:** Warum ist Explainable AI (XAI) für den Einsatz von Machine Learning Modellen in Unternehmen wichtig?

**Richtige Antwort:** XAI hilft, Modelle und deren Entscheidungen nachvollziehbar zu machen

Begründung: XAI erhöht die Transparenz und Akzeptanz von ML-Modellen.

## Musterlösung: Offene Fragen

**Offene Frage 1:** Nennen und erläutern Sie die sechs Phasen des QUA³CK-Prozessmodells anhand eines Beispiels.

Die sechs Phasen des QUA³CK-Prozessmodells:
- **Q**uestion: Definition des Business-Problems, z.B. Vorhersage von Immobilienpreisen.
- **U**nderstand: Datenexploration, z.B. Analyse von Preisverteilungen.
- **A**cquire & Clean: Datenbeschaffung und -bereinigung, z.B. Entfernen von Ausreißern.
- **A**nalyze: Modellierung, z.B. Training eines Random Forests.
- **A**pp: Entwicklung einer App, z.B. Streamlit-Dashboard zur Preisvorhersage.
- **C**onclusion & **K**ommunikation: Dokumentation und Präsentation der Ergebnisse.

**Offene Frage 2:** Beschreiben Sie einen vollständigen Machine-Learning-Workflow in Python von der Datenbeschaffung bis zur Evaluation.

Ein vollständiger ML-Workflow in Python:
- Datenimport mit Pandas (z.B. pd.read_csv)
- EDA: describe(), info(), Visualisierungen
- Feature Engineering: neue Features, One-Hot-Encoding
- Modelltraining mit Scikit-learn (fit)
- Evaluation: Accuracy, Confusion Matrix, Cross-Validation

**Offene Frage 3:** Was versteht man unter Data Leakage, wie erkennt man es und wie kann es verhindert werden?

Data Leakage:
- Bedeutung: Informationen aus Testdaten gelangen ins Training.
- Erkennung: Unerwartet hohe Testgenauigkeit, Feature-Analyse.
- Vermeidung: Strikte Trennung von Trainings- und Testdaten, Feature-Auswahl nur auf Trainingsdaten.

**Offene Frage 4:** Beschreiben Sie die wichtigsten Schritte und Herausforderungen bei der Entwicklung einer Streamlit-App für Machine Learning.

Entwicklung einer Streamlit-App:
- Daten laden und vorverarbeiten
- Modell laden/trainieren
- Benutzeroberfläche mit Streamlit-Widgets
- Vorhersagen anzeigen
- Herausforderungen: Performance, Usability, Deployment

**Offene Frage 5:** Vergleichen Sie Random Forests und KNN hinsichtlich Funktionsweise, Vor- und Nachteilen.

Random Forests vs. KNN:
- Random Forests: robust, weniger Overfitting, interpretierbar, aber komplexer.
- KNN: einfach, keine Trainingsphase, aber speicherintensiv und langsam bei großen Daten.

**Offene Frage 6:** Wie unterstützt MLflow das Experiment-Tracking im Machine Learning und welche Vorteile bietet es?

MLflow Experiment-Tracking:
- Experimente mit verschiedenen Parametern/Modellen anlegen
- Ergebnisse und Artefakte speichern
- Vorteile: Nachvollziehbarkeit, Vergleichbarkeit, Reproduzierbarkeit

**Offene Frage 7:** Beschreiben Sie, wie Transfer Learning im Bereich Computer Vision eingesetzt wird.

Transfer Learning für Computer Vision:
- Vortrainiertes Modell (z.B. ResNet) laden
- Eigene Daten für Feintuning nutzen
- Training, Evaluation, Deployment als App

**Offene Frage 8:** Wie werden Cross-Validation und Hyperparameter-Tuning in der Praxis kombiniert und umgesetzt?

Cross-Validation & Hyperparameter-Tuning:
- Cross-Validation für stabile Bewertung
- Grid Search/Random Search für Hyperparameter
- Praktisch: Pipeline in Scikit-learn nutzen

**Offene Frage 9:** Wie kann die Explorative Datenanalyse (EDA) zur Erkennung und Behandlung von Ausreißern und Fehlern genutzt werden?

EDA zur Fehler- und Ausreißererkennung:
- Visualisierungen (Boxplots, Scatterplots)
- Statistische Kennzahlen
- Maßnahmen: Entfernen, Imputieren, Transformation

**Offene Frage 10:** Welche Rolle spielt Explainable AI (XAI) für die Akzeptanz von Machine Learning Modellen?

Explainable AI (XAI) und Akzeptanz:
- XAI macht Modelle nachvollziehbar (z.B. SHAP, LIME)
- Erhöht Vertrauen und Akzeptanz bei Anwendern
- Beispiel: Feature-Importances visualisieren
