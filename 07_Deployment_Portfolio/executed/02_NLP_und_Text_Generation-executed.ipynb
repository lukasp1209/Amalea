```python
import random
from typing import List

def simple_sentiment(text: str):
    pos = sum(text.lower().count(w) for w in ['good','great','love','happy','nice'])
    neg = sum(text.lower().count(w) for w in ['bad','hate','sad','angry'])
    score = pos - neg
    if score>0: return 'POSITIVE', min(0.5+0.1*score,0.99)
    if score<0: return 'NEGATIVE', min(0.5+0.1*abs(score),0.99)
    return 'NEUTRAL', 0.55

def simple_generate(prompt: str, max_length: int=80):
    random.seed(42)
    suffix = ' '.join(['demo']*5)
    return (prompt + ' ' + suffix)[:max_length]

text = 'I love this course, it is good and fun'
label, conf = simple_sentiment(text)
print('Sentiment:', label, conf)
print('Generation:', simple_generate('Once upon a time'))
```
